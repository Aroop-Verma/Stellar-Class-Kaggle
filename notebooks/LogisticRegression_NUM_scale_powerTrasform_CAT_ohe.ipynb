```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import seaborn as sns
import joblib
```

```python
WORK_DIR="/home/magnolia/DataScience/Stellar_Class"
TRAIN_PATH=Path(WORK_DIR,"data/train.csv")
TEST_PATH=Path(WORK_DIR, "data/test.csv")

MODEL_DIR=Path(WORK_DIR,'models')
```

```python
df=pd.read_csv(TRAIN_PATH).set_index('id')
target=df['class']
df=df.drop('class', axis=1)
features=df.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PowerTransformer, StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
NUM_COLS=df.select_dtypes(['float', 'int']).columns
CAT_COLS=df.select_dtypes('object').columns

NUM_PIPE=Pipeline([('SCALER', StandardScaler()), ('POWER_TRANSFORM', 
                                                  PowerTransformer(method='yeo-johnson'))])
CAT_PIPE=Pipeline([('ENCODER', OneHotEncoder())])

transform=ColumnTransformer([('NUM_PIPE', NUM_PIPE, NUM_COLS), ('CAT_PIPE', CAT_PIPE, CAT_COLS)])
```

```python
X_t=transform.fit_transform(df)
```

```python
features_X_t=transform.get_feature_names_out()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import balanced_accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X_t, target)

LR=LogisticRegression()
LR.fit(X_train, y_train)
y_pred=LR.predict(X_test)
balanced_accuracy_score(y_test, y_pred)
```

```python
joblib.dump(LR, Path(MODEL_DIR,"LogisticRegression_NUM_scale_powerTrasform_CAT_ohe.joblib"))
```
