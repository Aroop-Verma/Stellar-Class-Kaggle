```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import seaborn as sns
import joblib
```

```python
WORK_DIR="/home/magnolia/DataScience/Stellar_Class"
TRAIN_PATH=Path(WORK_DIR,"data/train.csv")
TEST_PATH=Path(WORK_DIR, "data/test.csv")

MODEL_DIR=Path(WORK_DIR,"models")
```

```python
df=pd.read_csv(TRAIN_PATH).set_index('id')
target=df['class']
df=df.drop('class', axis=1)
features=df.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
# drop CAT Cols
df_NUM=df.select_dtypes(['float', 'int'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_NUM, target, stratify=target)
LR=LogisticRegression(n_jobs=-1, max_iter=2000)
LR.fit(X_train, y_train)
```

```python
from sklearn.metrics import balanced_accuracy_score
y_pred=LR.predict(X_test)
balanced_accuracy_score(y_test, y_pred)
```

```python
joblib.dump(LR, Path(MODEL_DIR, 'LogisticRegression_NUM_COLS_NO_TRANSFORMATION.joblib'))
```
