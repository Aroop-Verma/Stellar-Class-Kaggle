```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

class Engineer:

    def __init__(self, df: pd.DataFrame, target: str, id: str = 'id'):
        self.df         = df.copy()
        self.X          = df.set_index(id).drop(target, axis = 1).to_numpy()
        self.target     = df[target]
        self.shape      = df.shape
        self.features   = df.columns

        self.X_train    = self.X_test = None
        self.y_train    = self.y_test = None
        
        self.scaler_    = {}
        self.encoder_   = {}

        self.metrics    = {}


    def split(self, test_size: float = 0.3) -> "Engineer":
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(self.X, 
                                                                                self.target, 
                                                                                test_size=test_size, 
                                                                                random_state=10)
        return self
    
    def minmax(self, columns: list) -> "Engineer":
        for col in columns:
            scaler            = MinMaxScaler()
            self.scaler_[col] = scaler.fit_transform(self.X)

    def get_df(self) -> "Engineer":
        return self.df
```
