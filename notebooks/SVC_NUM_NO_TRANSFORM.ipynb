```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import joblib
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
WORK_DIR="/home/magnolia/DataScience/Stellar_Class"
TRAIN_PATH=Path(WORK_DIR,"data/train.csv")
TEST_PATH=Path(WORK_DIR, "data/test.csv")
MODEL_DIR=Path(WORK_DIR,"models")
```

```python
model=joblib.load(Path(MODEL_DIR, 'SVC_NO_CAT_COLS.joblib'))
```

```python
X=pd.read_csv(TEST_PATH).set_index('id').select_dtypes(['int', 'float']).drop(['alpha', 'delta'], axis=1)
```

```python
predictions=model.predict(X)
```

```python
predictions
```

```python
target_map={0: "GALAXY", 1: "QSAR", 2: "STAR"}
```

```python
submission=pd.DataFrame({'id':X.index, 'class':predictions}).set_index('id')
submission['class'].map(target_map).to_csv(Path(WORK_DIR,'submission','submission.csv'))
```
